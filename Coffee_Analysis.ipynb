# **COFFEE ANALYSIS**
Nescafe wants to understand their shoppers and their basic characteristics. They want an analysis of consumer characteristics, preferential data, and comparisons with Starbucks shoppers.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Consumers Characteristics**

```python
df = pd.read_excel("/content/Consumer Characteristics.xlsx", header= 3)
df.head(5)
```

```python
# Who are the Nescafe Buyers?
df['% of Nescafe Shoppers'] = pd.to_numeric(df['% of Nescafe Shoppers'], errors='coerce')
df.sort_values('% of Nescafe Shoppers', ascending=False).head(5)
```

```python
#Who are Starbucks buyer?
df['% of Starbucks Shoppers'] = pd.to_numeric(df['% of Starbucks Shoppers'], errors='coerce')
df.sort_values('% of Starbucks Shoppers', ascending=False).head(5)
```

```python
#Compare Nescafe vs Starbucks (differences)
df['Difference (Nescafe - Starbucks)'] = df['% of Nescafe Shoppers'] - df['% of Starbucks Shoppers']

# See top differences
df.sort_values('Difference (Nescafe - Starbucks)', ascending=False).head(10)   # Where Nescafe stronger
df.sort_values('Difference (Nescafe - Starbucks)').head(5)                   # Where Starbucks stronger

# Show top differences (positive = Nescafe stronger, negative = Starbucks stronger)
df[['Population Characteristics','Element','% of Nescafe Shoppers','% of Starbucks Shoppers','Difference (Nescafe - Starbucks)']].head(10)
```

# **Visualize Comparison (Example: Age Brackets, Income $, Education)**

```python

age_df = df[df['Population Characteristics'] == 'Age (Brackets)']

# Bar chart
plt.figure(figsize=(8,5))
sns.barplot(data=age_df, x='Element', y='% of Nescafe Shoppers', color='red', label='Nescafe')
sns.barplot(data=age_df, x='Element', y='% of Starbucks Shoppers', color='green', alpha=0.6, label='Starbucks')
plt.xticks(rotation=45)
plt.title("Age Distribution: Nescafe vs Starbucks Shoppers")
plt.ylabel("Percentage")
plt.legend()
plt.show()
```

```python
income_df = df[df['Population Characteristics'] == 'Income $']
plt.figure(figsize=(8,5))
sns.barplot(data=income_df, x='Element', y='% of Nescafe Shoppers', color='red', label='Nescafe')
sns.barplot(data=income_df, x='Element', y='% of Starbucks Shoppers', color='green', alpha=0.6, label='Starbucks')

plt.title("Nescafe vs Starbucks - Income Bracket Comparison")
plt.ylabel("Percentage of Shoppers")
plt.xticks(rotation=20)
plt.legend()
plt.show()
```

```python
Edu_df = df[df['Population Characteristics'] == 'Education']
plt.figure(figsize=(8,5))
sns.barplot(data=Edu_df, x='Element', y='% of Nescafe Shoppers', color='red', label='Nescafe')
sns.barplot(data=Edu_df, x='Element', y='% of Starbucks Shoppers', color='green', alpha=0.6, label='Starbucks')

plt.title("Nescafe vs Starbucks - Education Comparison")
plt.ylabel("Percentage of Shoppers")
plt.xticks(rotation=20)
plt.legend()
plt.show()
```

```python
hh_df = df[df['Population Characteristics'] == 'Household Size']

plt.figure(figsize=(8,5))
sns.barplot(data=hh_df, x='Element', y='% of Nescafe Shoppers', color='red', label='Nescafe')
sns.barplot(data=hh_df, x='Element', y='% of Starbucks Shoppers', color='green', alpha=0.6, label='Starbucks')
plt.title("Household Size: Nescafe vs Starbucks")
plt.ylabel("Percentage")
plt.legend()
plt.show()
```

```python
# Filter rows for Age (Generation)
gen_df = df[df['Population Characteristics'] == 'Age (Generation)']

# Select only useful columns
print(gen_df[['Element','% of Nescafe Shoppers','% of Starbucks Shoppers']])
```

```python
#Income Brackets
income_df = df[df['Population Characteristics']=='Income Bucket']
print(income_df[['Element','% of Nescafe Shoppers','% of Starbucks Shoppers']])
```

# **Consumer Preferential Data**

```python
import pandas as pd
data=pd.read_excel('/content/Consumers Preferential data.xlsx', header=3)
```

```python
data.head(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
#Convert Numeric Columns (to avoid string errors)
cols_to_convert = [
    '% of Starbucks Shoppers', '% of All Shoppers', 'Index',
    '% of Nescafe Shoppers'
]

data[cols_to_convert] = data[cols_to_convert].apply(pd.to_numeric, errors='coerce')
data.head(5)
```

```python
# Calculate difference (Starbucks - Nescafe)
data["Difference"] = data['% of Starbucks Shoppers'] - data['% of Nescafe Shoppers']

# Show largest differences
comparison = data[['Consumer\'s Preference','Header','Survey Question','Response',
                 '% of Starbucks Shoppers','% of Nescafe Shoppers','Difference']] \
             .sort_values(by="Difference", ascending=False)

print(comparison.head(10))  # Starbucks > Nescafe
print(comparison.tail(10))  # Nescafe > Starbucks
```

```python
import matplotlib.pyplot as plt

# Filter for Advertising category
ads = data[data["Consumer's Preference"]=="Advertising"]

plt.figure(figsize=(10,6))
plt.barh(ads['Response'], ads['% of Starbucks Shoppers'], color='green', alpha=0.6, label="Starbucks")
plt.barh(ads['Response'], ads['% of Nescafe Shoppers'], color='brown', alpha=0.6, label="Nescafe")

plt.xlabel("Percentage of Shoppers")
plt.title("Advertising Touchpoints: Starbucks vs Nescafe")
plt.legend()
plt.show()
```

```python
#Advertising Preferences
ads_df = data[data["Consumer's Preference"]=="Advertising"]
print(ads_df[['Response','% of Nescafe Shoppers','% of Starbucks Shoppers']])
```
